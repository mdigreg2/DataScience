<h3>Pandas' series type can be used to make a flexible python list. Data referencing works like it does with numpy arrays. Named variables seem to be references to the same data, which is why you can cange one reference by changing another. This saves on space and computing resources. See the last three lines for this example</h3>

```python
# An important part of pandas is "series" They are built off of python lists bu are much more flexible

import pandas as pd
import numpy as np

s = pd.Series(np.random.randn(4), name = 'daily returns')
print(s)

s.index = ['AMAZN', 'GOOGL','WISA','CITI']
print(s)

print(s.describe())
print(s * 100)
print(s['AMAZN'])
s['AMAZN'] = 0
print(s)
b = s
b['GOOGL'] = 100
print(s)
```

<h3>Individual features can be graphed and new ones created by simply labeling another column and populating it</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline

df = pd.read_csv('../data/Demographic_Statistics_By_Zip_Code.csv')

df = df.loc[df.index[0:10], ['JURISDICTION NAME', 'COUNT MALE', 'COUNT FEMALE', 'COUNT CITIZEN STATUS TOTAL']]
df.columns = 'Jurisdiction', 'Male', 'Female', 'Total'
df['Ratio'] = df['Male'] / df['Female']
df['Male'].plot(kind = 'bar')
pt.show()
```

<h3>Data can also be sorted within the program using pandas' sort_values() method</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline

df = pd.read_csv('../data/Demographic_Statistics_By_Zip_Code.csv')

df = df.loc[df.index[0:10], ['JURISDICTION NAME', 'COUNT MALE', 'COUNT FEMALE', 'COUNT CITIZEN STATUS TOTAL']]
df.columns = 'Jurisdiction', 'Male', 'Female', 'Total'
df['Ratio'] = df['Male'] / df['Female']
df = df.sort_values(by = 'Male', ascending=False)
df['Male'].plot(kind = 'bar')
pt.show()
```

<h3>Using World Bank debt data: </h3>

```python
import pandas as pd
import matplotlib.pyplot as pt
import requests
%matplotlib inline


# get data and read into the file gd.xls
url = "http://api.worldbank.org/v2/en/indicator/gc.dod.totl.gd.zs?downloadformat=excel"
r = requests.get(url)
with open('gd.xls', 'wb') as output:
    output.write(r.content)

# put the data into a data frame
govtDebt = pd.read_excel('gd.xls', sheet_name = 'Data', skiprows = 3, index_col = 1)
govtDebt = govtDebt.transpose()
govtDebt = govtDebt[['AUS', 'USA']]
govtDebt = govtDebt[30:]
govtDebt.plot(lw = 2, alpha = .6)
pt.show()


# the .transpose() function reverses rows and columns. i.e. rows become columns and columns become rows

```
