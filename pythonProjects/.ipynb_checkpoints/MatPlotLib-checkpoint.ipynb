<h3>Matplotlib is cool, you can use it matlab style or python object oriented style. It also handles 3D graphs well</h3>

```python
# Matlab way
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(1,10, 50)
y = np.sin(x)

plt.plot(x,y,'-r',linewidth = 2, label = 'sin(x)')
plt.title('matlab style is x,y then plot them')
plt.legend()
plt.show()
```

```python
# The python way
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


# Fig is like a zone in a graph   it basically holds our graph 
# fig is a figure instance, like a blank canvas
# ax is an axes subplot i.e a frame for painting in 
# plot is a method of ax
fig, ax = plt.subplots()
x = np.linspace(1,10, 50)
y = np.sin(x)


# alpha changes transparency and makes the line look smoother
ax.plot(x,y, '-b', linewidth = 2, label = "$y=\sin(x)$" , alpha = .6)
ax.set_yticks([-1,0,1])
ax.set_title('Python style plot')
plt.legend(loc = 'lower left')
plt.show()
```

```python
# Multiple plots can share the same figure and axes
import matplotlib.pyplot as plt
import numpy as np

a = (9,6,3)
x = np.linspace(0,(np.pi*5),200)
y = np.sin(x)
fig,ax = plt.subplots()
for index,coeff in enumerate(a):
    currentLabel = f'$\mu = {a[index]}$'
    ax.plot(x,(np.sin(x + (9*coeff))), linewidth = 2, alpha = .6, label = currentLabel)
plt.legend(loc = 'upper right')
plt.show()
```

```python
# if you want multiple graphs on one output
import matplotlib.pyplot as plt
import numpy as np

rows, columns = 3,2
x = np.linspace(0,10,(400.0/np.pi))
y = np.sin(x)
fig,axes = plt.subplots(rows,columns,figsize = (10,12))
for i in range(rows):
    for j in range(columns):
        currentLabel = 1
        axes[i,j].plot(x,y,linewidth = 2, alpha = .6,label = currentLabel)
        currentLabel+=1
plt.legend(loc = 'upper right')
plt.show()
        
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import numpy as np

def f(x,y):
    return np.cos(x**2+y**2) / (1 + x**2+y**2)

xgrid = np.linspace(-3,3,50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(x,
                y,
                f(x,y),
                rstride = 2, cstride = 2,
                cmap = cm.jet,
                alpha = .7,
                linewidth = .25)
ax.set_zlim(-.5,1.0)
plt.show()
```

<h3>Plot the function
f(x)=cos(πθx)exp(−x)

over the interval [0,5]
for each θ

in np.linspace(0, 2, 10)

Place all the curves in the same figure</h3>


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

theta = np.linspace(0,2,10)
fig,ax = plt.subplots()
x = np.linspace(0,5,100)
for index,i in enumerate(theta):
    y = np.cos(np.pi*i*x) * np.exp(-x)
    ax.plot(x,y, linewidth = 1, alpha = .7)
plt.show()
```
