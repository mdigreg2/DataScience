<h3>This computes a factorial. Mostly for parctice with for loops in python and function definitions</h3>

```python
def factorial(n):
    answer = 1
    for i in range(1, n):
        answer = answer * i
    return answer

n = 10
print(factorial(n))
```

<h3>The binomial random variable Y∼Bin(n,p) represents the number of successes in n binary trials, where each trial succeeds with probability p

Without any import besides from numpy.random import uniform, write a function binomial_rv such that binomial_rv(n, p) generates one draw of Y

Hint: If U
is uniform on (0,1) and p∈(0,1), then the expression U < p evaluates to True with probability p</h3>

```python
from numpy.random import uniform

def binomial_rv(n, p):
    counter = 0
    Fcounter = 0
    for i in range(n):
        result = uniform(0,1)
        if result <= p:
            counter += 1
        else:
            Fcounter += 1
    return (counter)

print(binomial_rv(5, .50))
```

<h3>Compute an approximation to pi using Monte Carlo using no imports besides "import numpy as np"<br>Hint:<br><br>If U is a bivariate uniform random variable on the unit square (0,1)^2, then the probability that U lies in a subset B of (0,1)^2 is equal to the area of B
<br><br>If U1,…,Un are iid copies of U, then, as n gets large, the fraction that fall in B converges to the probability of landing in B
<br><br>For a circle, area = pi * radius^2 </h3>

```python
import numpy as np
# Monte carlo approximation of the average of np.random.uniform
# Using a basic interpretation of the process
# approx = (1/n)* SUM(f(xi)) from i = 0 to i = n
def monteCarlo(n):
    counter = 0
    totalTrials = n
    for i in range(totalTrials):
        trialX = np.random.uniform(0,100)
        counter += trialX
    average = (1/totalTrials)*counter
    return average

print(monteCarlo(10000))
```

```python
import numpy as np
# This uses the np.random.uniform method to generate a "square" of
# side length 1. It then determines if the mangitude of the vector
# created by trialX and trialY is <= 1 (i.e if they fit into a
# quarter circle with its center at a corner of the square) 
# The probability of the point landing within that circle is its area
# The radius of the circle is 1
# A = pi * r^2, however we are using a quarter circle so A = (pi/4)*r^2
# r = 1 so 4*A = pi  
# A = (number of trials within the circle)/(Total Trials)
def monteCarlo(n):
    totalTrials = n
    counter = 0
    for i in range(totalTrials):
        trialX = np.random.uniform(0,1)
        trialY = np.random.uniform(0,1)
        if ((trialY*trialY)+(trialX*trialX)) <= 1:
            counter += 1
    prob = (counter*4) / totalTrials
    return prob


print(monteCarlo(100000))
    
```

<h3>Your next task is to simulate and plot the correlated time series</h3>
x_t+1= αx_t+ε_t+1 wherex_0=0 and t=0,…,T
<br>

The sequence of shocks {ε_t}

 is assumed to be iid and standard normal
<br><br>
<h3>
In your solution, restrict your import statements to

<br>
Set T=200

and α=0.9</h3>


```python
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline

def generateTimeSeries(a, t):
    currentX = 0    
    xt = []
    for i in range(t+1):
        xt.append(currentX)
        currentX = a*currentX + np.random.randn()
        
    return xt

d0 = generateTimeSeries(0,200)
d8 = generateTimeSeries(.8,200)
d98 = generateTimeSeries(.98,200)
pt.plot(d0, label = "a = 0.0")
pt.plot(d8, label = "a = 0.8")
pt.plot(d98, label = "a = 0.98")
pt.legend()
pt.show()
```
